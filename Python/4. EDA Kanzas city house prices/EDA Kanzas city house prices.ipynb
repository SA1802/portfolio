
### 1.2 Изучите распределение стоимости недвижимости

```python
def distribution_and_boxplot(column):
    plt.figure(figsize = (15,9))
    
    # график распределения признака     
    plt.subplot(2,1,1)
    sns.histplot(df[column], kde = True)
    plt.axvline(x = df[column].median(), c = 'r', label = 'Медиана')
    plt.axvline(x = df[column].mean(), c = 'g', label = 'Среднее значение')
    plt.xticks([round(df[column].mean(), 1), round(df[column].median(), 1)], rotation = 90, fontsize = 8)
    plt.legend()
    plt.title(f"Распределение признака '{column}'", fontsize = 15)
    plt.ylabel('Частота', fontsize = 12)
    plt.xlabel('')
    plt.grid()
    
    # ящик с усами     
    plt.subplot(2,1,2)
    plt.grid()
    plt.boxplot(df[column], vert = False)
    plt.axvline(x = df[column].median(), c = 'r')
    plt.xlabel(column, fontsize = 12);
    
distribution_and_boxplot('price')
```

###### Вывод:
По графикам распределения стоимости недвижимости видно, что основная часть домов имеет примерную стоимость от 320 000 до 640 000 долларов. Стоимость почти всех домов укладывается в диапазон от 75 000 до около 1 100 000. 50% Домов имеют стоимость менее 450 000, отальные 50% - более 450 000. Средняя стоимость домов - 540 000.
Также по ящику с усами видно большое количестов выбросов, которым соответствуют дома с нетипично высокой стоимостью.

### Корреляционная матрица

```python
corr_matrix = df.corr()
corr_matrix[np.abs(corr_matrix) < 0.25] = 0
plt.figure(figsize = (14, 10))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
plt.title('Корреляционная матрица', fontsize = 15);
```

###### Вывод: 
По корреляционной матрице видно, что сильнее всего на стоимость недвижимости влияет внутренняя жилая площадь (sqft_living), так как при изменении на единицу данного признака стоимость прямопропорционально увеличится на 0,7. В то же время этот признак имеет очень сильную связь с другими признаками площади.

Также сильную прямую связь с площадью имеет признак grade, означающий общую оценку дома по шкале от 1 до 13. При этом он имеет сильную связь с sqft_living, видимо, чем больше внутренняя площадь дома, тем выше его общая оценка.

Сильное влияние на стоимость также имеет внутренняя жилая площадь над землей (sqft_above), но также этот признак имеет гораздо более высокую связь с sqft_living, что очевидно.

Признак "количество ванных" (bathrooms) тоже сильно влияет на стоимость, но опять-таки он гораздо сильнее связан с sqft_living.

Среди признаков, не связанных с площадью достаточно высокую связь имеет признак view.

Рассмотрим подробнее влияние на стоимость тех признаков, которые имеют самую сильную связь со стоимостью, но при этом не будем разбирать признаки, которые сильнее связаны между собой, чем со стоимостью.

### Диаграмма рассеяния стоимости и общей жилой площади

```python
plt.figure(figsize = (8, 10))
sns.jointplot(x = df['sqft_living'], y = df['price'], kind = 'reg')
plt.title('Диаграмма рассеяния стоимости и общей жилой площади', fontsize = 15, pad = 100, loc = 'right');
```

###### Вывод: 
По графику видна положительная корреляция между стоимостью и общей жилой площадью, то есть чем больше площадь, тем выше стоимость. Примечательно, что скопление точек очень четко ограничено сверху и снизу будто прямыми линиями, на основе чего можно сделать вывод, что для каждого объема жилой площади есть явный диапазон, в котором будет находиться стоимость. При этом чем больше стоимость, тем этот коридор шире.

### Влияние вида из дома на стоимость

```python
plt.figure(figsize = (15, 10))
sns.boxplot(x = df['view'], y = df['price'], showfliers=False)
plt.title('Влияние вида из дома на стоимость', fontsize = 15);
plt.xlabel('Вид из дома', fontsize = 12)
plt.ylabel('Стоимость', fontsize = 12);
```

###### Вывод:
Между стоимостью дома и видом также видна прямая связь (чем лучше вид из дома, тем больше стоимость). При этом заметно, что нижняя граница стоимости для каждой оценки вида примерно одинакова, в то время как верхняя граница повышается с улучшением вида.

### Влияние этажности на стоимость

Соотнося дома с одинаковым числом этажей, но различающиеся наличием подвала, можно сказать, что наличие подвала заметно повышает стоимость дома.

Так, например, явно видно наличие районов с более дорогими домами: 98039, 98004, 98112. В этих районах стоимость заметно выше, а также стоимость в них имеет самый большой разброс. Скорее всего, это дома близкие к деловому центру или, так как Сиэтл имеет выход к морю, расположенные у берега.

Есть районы и с дешевыми домами: группы индексов 98001-98003,  98030-98032 и 98178-98198 имеют примерно одинаково очень низкую цену. Вероятно, эти крупные районы, объединяющие несколько индексов, имеют плохую инфраструктуру или географическое положение (окраины, плохая экология и т.д.).
